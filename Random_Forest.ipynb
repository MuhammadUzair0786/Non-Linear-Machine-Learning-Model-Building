## Random Forest Algorithm

#### Random Forest Algorithm Solve Classification

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset =pd.read_csv(r'C:\Users\Muhammad Uzair\Downloads\Datasets\Iris.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.drop('Id',axis= 1,inplace=True)
```

```python
x =dataset.iloc[:,:-1]
y =dataset['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test ,y_train, y_test =train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc =RandomForestClassifier(n_estimators=50, criterion='entropy',max_depth=6)
rfc.fit(x_train,y_train)
```

```python
print(f'Train Score :{rfc.score(x_train,y_train)*100}')
print(f'Test Score :{rfc.score(x_test,y_test)*100}')
```

```python
from sklearn.metrics import confusion_matrix
cf =confusion_matrix(y_test,y_pred=rfc.predict(x_test))
```

```python
sns.heatmap(cf,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
rfc.predict([[5.1,3.5,1.4,0.2]]) #  5.1	3.5	1.4	0.2	Iris-setosa
```

```python
rfc.predict([[5.9,3.0,5.1,1.8]]) # 5.9	3.0	5.1	1.8	Iris-virginica
```

```python

rfc.predict([[5.5,2.5,4,1.3]]) # 5.5	2.5	4	1.3	Iris-versicolor
```

#### Random Forest Algorithm Solve Regression

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset =pd.read_csv(r'C:\Users\Muhammad Uzair\Downloads\Datasets\Salary_dataset.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
x =dataset.iloc[:,:-1]
y =dataset['Salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test ,y_train, y_test =train_test_split(x,y,test_size=0.20,random_state=20)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr =RandomForestRegressor(n_estimators=10, criterion='squared_error',max_depth=5)
rfr.fit(x_train,y_train)
```

```python
print(f'Train Score :{rfr.score(x_train,y_train)*100}')
print(f'Test Score :{rfr.score(x_test,y_test)*100}')
```

```python
rfr.predict([[5]]) # 5	67939
```
